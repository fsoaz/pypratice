```python
'''Utilizando listas faça um programa que faça 5 perguntas para uma pessoa sobre um crime. As perguntas são:

    "Telefonou para a vítima?"
    "Esteve no local do crime?"
    "Mora perto da vítima?"
    "Devia para a vítima?"
    "Já trabalhou com a vítima?" O programa deve no final emitir uma classificação sobre a participação da pessoa no crime. Se a pessoa responder positivamente a 2 questões ela deve ser classificada como "Suspeita", entre 3 e 4 como "Cúmplice" e 5 como "Assassino". Caso contrário, ele será classificado como "Inocente".'''

'''Utilizando listas faça um programa que faça 5 perguntas para uma pessoa sobre um crime. As perguntas são:

    "Telefonou para a vítima?"
    "Esteve no local do crime?"
    "Mora perto da vítima?"
    "Devia para a vítima?"
    "Já trabalhou com a vítima?" O programa deve no final emitir uma classificação sobre a participação da pessoa no crime. Se a pessoa responder positivamente a 2 questões ela deve ser classificada como "Suspeita", entre 3 e 4 como "Cúmplice" e 5 como "Assassino". Caso contrário, ele será classificado como "Inocente".'''

todas_repostas = []
print("Digite (1) para Sim e (0) para Não\n")

def get_valid_input(prompt):
    while True:
        try:
            resposta = int(input(prompt))
            if resposta in [0, 1]:
                return resposta
            else:
                print("Por favor, digite (1) para Sim ou (0) para Não.")
        except ValueError:
            print("Entrada inválida! Digite apenas 0 ou 1.")

telefone = get_valid_input("Telefonou para a vítima? ")
local_crime = get_valid_input("Esteve no local do crime? ")
mora_perto = get_valid_input("Mora perto da vítima? ")
divida = get_valid_input("Devia para a vítima")
ja_trabalhou = get_valid_input("Já trabalhou com a vítima? ")

resposta = {
    "Telefonou para a vítima": telefone,
    "Esteve no local do crime": local_crime,
    "Mora perto da vítima": mora_perto,
    "Já trabalhou com a vítima": ja_trabalhou,
    "Devia para a vítima": divida}
todas_repostas.append(resposta)

num_respostas_positivas = sum(resposta.values())

if num_respostas_positivas == 2:
    print("\nClassificação: Suspeito")
elif num_respostas_positivas == 3 or num_respostas_positivas == 4:
    print("\nClassificação: Cúmplice")
elif num_respostas_positivas == 0 or num_respostas_positivas == 1:
    print("\nClassificação: Inocente")
elif num_respostas_positivas > 4:
    print("\nClassificação: Assassino!")
else:
    print("\nClassificação: Inocente")


print("\nRespostas coletadas:")
for key, value in resposta.items():
    print(f"{key}: {'Sim' if value == 1 else 'Não'}")

```

```python
'''Utilize uma lista para resolver o problema a seguir. Uma empresa paga seus vendedores com base em comissões. O vendedor recebe $200 por semana mais 9 por cento de suas vendas brutas daquela semana. Por exemplo, um vendedor que teve vendas brutas de $3000 em uma semana recebe $200 mais 9 por cento de $3000, ou seja, um total de $470. Escreva um programa (usando um array de contadores) que determine quantos vendedores receberam salários nos seguintes intervalos de valores:

    $200 - $299
    $300 - $399
    $400 - $499
    $500 - $599
    $600 - $699
    $700 - $799
    $800 - $899
    $900 - $999
    $1000 em diante 

Desafio: Crie ma fórmula para chegar na posição da lista a partir do salário, sem fazer vários ifs aninhados. '''

# Lista para armazenar vendedores
vendedores = []

# Lista de contadores para cada faixa salarial
faixas = [0] * 9  # 9 faixas salariais

# Definição dos intervalos salariais
intervalos = [
    "$200 - $299", "$300 - $399", "$400 - $499", "$500 - $599",
    "$600 - $699", "$700 - $799", "$800 - $899", "$900 - $999",
    "$1000 ou mais"
]

i = 0  # Contador de vendedores
while True:
    nome = input(f"Nome do {i+1}º vendedor: ")
    vendas = float(input(f"Vendas do {i+1}º vendedor: "))

    # Cálculo do salário do vendedor
    salario = 200 + (vendas * 0.09)

    # Adicionar vendedor à lista
    vendedores.append({"nome": nome, "salario": salario})

    # Perguntar se deseja continuar
    continuar = input("Deseja adicionar outro vendedor? (s/n): ").strip().lower()
    if continuar != 's':
        break

    i += 1  # Incrementar contador de vendedores

# Processamento das faixas salariais
for vendedor in vendedores:
    salario = vendedor["salario"]

    # Determinar faixa usando indexação matemática
    indice = min(int((salario - 200) // 100), 8)
    faixas[indice] += 1  # Incrementar contador da faixa correspondente

# Exibição dos resultados
print("\nDistribuição dos vendedores por faixa salarial:")
for i, intervalo in enumerate(intervalos):
    print(f"{intervalo}: {faixas[i]} vendedor(es)")
```

```python
'''Em uma competição de salto em distância cada atleta tem direito a cinco saltos. O resultado do atleta será determinado pela média dos cinco valores restantes. Você deve fazer um programa que receba o nome e as cinco distâncias alcançadas pelo atleta em seus saltos e depois informe o nome, os saltos e a média dos saltos. O programa deve ser encerrado quando não for informado o nome do atleta. A saída do programa deve ser conforme o exemplo abaixo: 

Atleta: Rodrigo Curvêllo
 
Primeiro Salto: 6.5 m
Segundo Salto: 6.1 m
Terceiro Salto: 6.2 m
Quarto Salto: 5.4 m
Quinto Salto: 5.3 m

Resultado final:
Atleta: Rodrigo Curvêllo
Saltos: 6.5 - 6.1 - 6.2 - 5.4 - 5.3
Média dos saltos: 5.9 m'''

i = 0
atletas = []
salto_ordem = ["Primeiro", "Segundo", "Terceiro", "Quarto", "Quinto"]

while True:
    atleta = input(f"Nome do(a) {i+1}º atleta: ")

    # Collect jump distances
    saltos = []
    for x in range(5):
        distancia_salto = float(input(f"Distância do {x+1}º salto: "))
        saltos.append(distancia_salto)

    # Calculate the average jump distance
    media = sum(saltos) / len(saltos)

    # Store athlete's data including the average
    atletas.append({"nome": atleta, "saltos": saltos, "media": media})

    continuar = input("Deseja adicionar outro atleta? (s/n): ").strip().lower()
    if continuar != 's':
        break

    i += 1  # Increment the athlete counter

# Display collected data
print("\nResultado final:\n")
for atleta in atletas:
    print(f"Atleta: {atleta['nome']}")
    for i, salto in enumerate(atleta['saltos']):
        print(f"{salto_ordem[i]} Salto: {salto:.1f} m")
    
    # Print summary of results
    saltos_formatados = " - ".join(f"{s:.1f}" for s in atleta['saltos'])
    print(f"Saltos: {saltos_formatados}")
    print(f"Média dos saltos: {atleta['media']:.1f} m\n")
```

```python

'''Uma grande emissora de televisão quer fazer uma enquete entre os seus telespectadores para saber qual o melhor jogador após cada jogo. Para isto, faz-se necessário o desenvolvimento de um programa, que será utilizado pelas telefonistas, para a computação dos votos. Sua equipe foi contratada para desenvolver este programa, utilizando a linguagem de programação C++. Para computar cada voto, a telefonista digitará um número, entre 1 e 23, correspondente ao número da camisa do jogador. Um número de jogador igual zero, indica que a votação foi encerrada. Se um número inválido for digitado, o programa deve ignorá-lo, mostrando uma breve mensagem de aviso, e voltando a pedir outro número. Após o final da votação, o programa deverá exibir:

  a. O total de votos computados;
  a. Os númeos e respectivos votos de todos os jogadores que receberam votos;
  a. O percentual de votos de cada um destes jogadores;
  a. O número do jogador escolhido como o melhor jogador da partida, juntamente com o número de votos e o percentual de votos dados a ele.

   Observe que os votos inválidos e o zero final não devem ser computados como votos. O resultado aparece ordenado pelo número do jogador. O programa deve fazer uso de arrays.
   O programa deverá executar o cálculo do percentual de cada jogador através de uma função. Esta função receberá dois parâmetros: o número de votos de um jogador e o total de votos. A função calculará o percentual e retornará o valor calculado.
   Abaixo segue uma tela de exemplo. O disposição das informações deve ser o mais próxima possível ao exemplo. Os dados são fictícios e podem mudar a cada execução do programa.
   Ao final, o programa deve ainda gravar os dados referentes ao resultado da votação em um arquivo texto no disco, obedecendo a mesma disposição apresentada na tela.'''

# Removed unused variable 'total_votos'
votos_jogador = []
def get_valid_input(prompt):
    while True:
        try:
            numero = int(input(prompt))
            if numero >= 0:
                return numero
            else:
                print("Por favor, digite um número maior que 0 ou 0 para sair.")
        except ValueError:
                print("Por favor, digite um número entre 1 e 23, ou 0 para encerrar.")


while True:
    votos = get_valid_input("Informe o número do jogador: (Digite 0 para sair): ")

    if votos == 0:
        break

    votos_jogador.append(votos)
from collections import Counter

print(f"Foram computados {len(votos_jogador)} votos")
separator = "=" * (len("Jogador") + len("Votos") + len("%") + 6)  # Adjust for column widths and spaces
print(separator + "\n")
# Count votes for each player
votos_contados = Counter(votos_jogador)

print("Jogador| Votos|  %")
print("========================")

# Calculate and display results
for jogador, votos in sorted(votos_contados.items()):
    percentual = (votos / len(votos_jogador)) * 100
    print(f"{jogador:^7}|{votos:^5} |{percentual:>5.2f}%")

# Determine the best player
melhor_jogador, max_votos = max(votos_contados.items(), key=lambda x: x[1])
percentual_melhor = (max_votos / len(votos_jogador)) * 100

print("\nO melhor jogador foi o número", melhor_jogador, "com", max_votos, 
      f"votos, correspondendo a {percentual_melhor:.2f}% do total de votos.")

```

```python
'''Uma empresa de pesquisas precisa tabular os resultados da seguinte enquete feita a um grande quantidade de organizações:
"Qual o melhor Sistema Operacional para uso em servidores?"

As possíveis respostas são:

1- Windows Server
2- Unix
3- Linux
4- Netware
5- Mac OS
6- Outro

Você foi contratado para desenvolver um programa que leia o resultado da enquete e informe ao final o resultado da mesma. O programa deverá ler os valores até ser informado o valor 0, que encerra a entrada dos dados. Não deverão ser aceitos valores além dos válidos para o programa (0 a 6). Os valores referentes a cada uma das opções devem ser armazenados num vetor.
Após os dados terem sido completamente informados, o programa deverá calcular a percentual de cada um dos concorrentes e informar o vencedor da enquete. O formato da saída foi dado pela empresa, e é o seguinte:

Sistema Operacional Votos   %
------------------- -----   ---
Windows Server       1500   17%
Unix             3500   40%
Linux            3000   34%
Netware           500    5%
Mac OS            150    2%
Outro             150    2%
------------------- -----
Total            8800

O Sistema Operacional mais votado foi o Unix, com 3500 votos, correspondendo a 40% dos votos.'''

# Definindo as opções de sistema operacional
options = {
    1: "Windows Server",
    2: "Unix",
    3: "Linux",
    4: "Netware",
    5: "Mac OS",
    6: "Outro"
}
# Inicializando o vetor de votos
votes = []# Inicializando o contador de votos
total_votes = 0
# Lendo os votos
while True:
    try:
        vote = int(input("Digite o número correspondente ao seu voto (0 para encerrar): "))
        if vote == 0:
            break
        elif vote in options:
            votes.append(vote)
            total_votes += 1
        else:
            print("Opção inválida. Tente novamente.")
    except ValueError:
        print("Entrada inválida. Por favor, digite um número.")
# Contando os votos
vote_count = {option: 0 for option in options}
for vote in votes:
    vote_count[vote] += 1

#Show the results
print("\nSistema Operacional Votos   %")
print("------------------- -----   ---")
for option, name in options.items():
    count = vote_count[option]
    percentage = (count / total_votes) * 100 if total_votes > 0 else 0
    print(f"{name:<22} {count:<5} {percentage:5.0f}%")
print("----------------------    -----")
print(f"Total                    {total_votes}")

# Determinando o vencedor
max_votes = max(vote_count.values())
winners = [name for option, name in options.items() if vote_count[option] == max_votes]

if len(winners) > 1:
    print(f"O(s) Sistema(s) Operacional(is) mais votado(s) foi(ram): {', '.join(winners)} com {max_votes} votos.")
else:
    winner = winners[0]
    print(f"O Sistema Operacional mais votado foi o {winner} com {max_votes} votos, correspondendo a {max_votes / total_votes * 100:.0f}% dos votos.")
```

```python
'''
    As Organizações Tabajara resolveram dar um abono aos seus colaboradores em reconhecimento ao bom resultado alcançado durante o ano que passou. Para isto contratou você para desenvolver a aplicação que servirá como uma projeção de quanto será gasto com o pagamento deste abono.

  Após reuniões envolvendo a diretoria executiva, a diretoria financeira e os representantes do sindicato laboral, chegou-se a seguinte forma de cálculo:

 a.Cada funcionário receberá o equivalente a 20% do seu salário bruto de dezembro;
 a.O piso do abono será de 100 reais, isto é, aqueles funcionários cujo salário for muito baixo, recebem este valor mínimo;

 Neste momento, não se deve ter nenhuma preocupação com colaboradores com tempo menor de casa, descontos, impostos ou outras particularidades. Seu programa deverá permitir a digitação do salário de um número indefinido (desconhecido) de salários. Um valor de salário igual a 0 (zero) encerra a digitação. Após a entrada de todos os dados o programa deverá calcular o valor do abono concedido a cada colaborador, de acordo com a regra definida acima. Ao final, o programa deverá apresentar:

 a. O salário de cada funcionário, juntamente com o valor do abono;
 a. O número total de funcionário processados;
 a. O valor total a ser gasto com o pagamento do abono;
 a. O número de funcionário que receberá o valor mínimo de 100 reais;
 a. O maior valor pago como abono;

 A tela abaixo é um exemplo de execução do programa, apenas para fins ilustrativos. Os valores podem mudar a cada execução do programa.

Projeção de Gastos com Abono
============================ 
 
Salário: 1000
Salário: 300
Salário: 500
Salário: 100
Salário: 4500
Salário: 0
 
Salário    - Abono     
R$ 1000.00 - R$  200.00
R$  300.00 - R$  100.00
R$  500.00 - R$  100.00
R$  100.00 - R$  100.00
R$ 4500.00 - R$  900.00
 
Foram processados 5 colaboradores
Total gasto com abonos: R$ 1400.00
Valor mínimo pago a 3 colaboradores
Maior valor de abono pago: R$ 900.00'''

funcionarios = []
abonos = []
total_abono = 0
colaboradores_minimo = 0
maior_abono = 0

def get_valid_input(prompt):
    while True:
        try:
            salario = float(input(prompt))
            if salario < 0:
                print("Por favor, insira um valor positivo.")
            else:
                return salario
        except ValueError:
            print("Entrada inválida. Por favor, insira um número.")

print("Projeção de Gastos com Abono")

while True:
    salario = get_valid_input("Informe o salário (Digite 0 para sair): ")
    if salario == 0:
        break

    funcionarios.append(salario)
    abono = max(100, salario * 0.2)
    abonos.append(abono)
    total_abono += abono
    if abono == 100:
        colaboradores_minimo += 1 #Contar colaboradores que recebem o mínimo
    maior_abono = max(maior_abono, abono)

#Exibir os resultados
print("\nSalário    - Abono")
for salario, abono in zip(funcionarios, abonos):
    print(f"R$ {salario:.2f} - R$ {abono:.2f}")

print(f"Foram processados {len(funcionarios)} colaboradores")
print(f"Total gasto com abonos: R$ {total_abono:,.2f}")
print(f"Valor mínimo pago a {colaboradores_minimo} colaboradores")
print(f"Maior valor de abono pago: R$ {maior_abono:,.2f}")
```

```python
'''Sua organização acaba de contratar um estagiário para trabalhar no Suporte de Informática, com a intenção de fazer um levantamento nas sucatas encontradas nesta área. A primeira tarefa dele é testar todos os cerca de 200 mouses que se encontram lá, testando e anotando o estado de cada um deles, para verificar o que se pode aproveitar deles.Foi requisitado que você desenvolva um programa para registrar este levantamento. O programa deverá receber um número indeterminado de entradas, cada uma contendo:
um número de identificação do mouse
o tipo de defeito:
a. necessita da esfera;
a. necessita de limpeza;
a.necessita troca do cabo ou conector;
a.quebrado ou inutilizado
Uma identificação igual a zero encerra o programa. Ao final o programa deverá emitir o seguinte relatório:

Quantidade de mouses: 100

Situação            Quantidade      Percentual
1- necessita da esfera          40             40%
2- necessita de limpeza         30             30%
3- necessita troca do cabo ou conector  15                     15%
4- quebrado ou inutilizado      15             15%'''

# Definindo as constantes
TIPOS_DE_DEFEITO = {
    1: "necessita da esfera",
    2: "necessita de limpeza",
    3: "necessita troca do cabo ou conector",
    4: "quebrado ou inutilizado"
}
contagem = {1: 0, 2: 0, 3: 0, 4: 0}
# Inicializando variáveis

total = int(input("Digite o número total de mouses testados: "))

for i in range(total):
    while True:
        try:
            defeito = int(input(f"Digite o tipo de defeito do {i + 1}º mouse (1-4): "))
            if defeito in TIPOS_DE_DEFEITO:
                contagem[defeito] += 1
                break
            else:
                print("Tipo de defeito inválido. Tente novamente.")
        except ValueError:
            print("Entrada inválida. Por favor, insira um número entre 1 e 4.")
# Exibindo o relatório
print("\nRELATÓRIO DE DEFEITOS")
print("Quantidade de mouses:", total)
print(f"{'Situação':40} {'Quantidade':>10} {'Percentual':>12}")
for codigo in sorted(TIPOS_DE_DEFEITO):
    qtd = contagem[codigo]
    percentual = (qtd / total) * 100 if total > 0 else 0
    print(f"{TIPOS_DE_DEFEITO[codigo]:40} {qtd:10} {percentual:11.2f}%")
```

```python
'''Faça um programa que simule um lançamento de dados. Lance o dado 100 vezes e armazene os resultados
em um vetor . Depois, mostre quantas vezes cada valor foi conseguido.
Dica: use um vetor de contadores(1-6) e uma função para gerar numeros aleatórios, simulando os
lançamentos dos dados.
A função que gera números inteiros é randint(ini,fim) e gera um dos números do intervalo de ini até fim.
Para usar essa função, é necessário fazer uma importação de biblioteca de Python chamada random. Essa
importação é feita assim:
import random
Exemplo de utilização:
random.ranint(1,3)
pode gerar aleatoriamente os números 1, 2 ou 3, a cada vez que é executada.'''

from random import randint
import random

def lancar_dados(qtd_lancamentos, qtd_faces):
    frequencia = {i: 0 for i in range(1, qtd_faces + 1)}
    #Realiza os lançamentos
    for _ in range(qtd_lancamentos):
        resultado = random.randint(1, qtd_faces)
        frequencia[resultado] += 1
    return frequencia
    
#quantidade de lançamentos e quantidade de faces do dado
if __name__ == "__main__":
    qtd_lancamentos = int(input("Digite a quantidade de lançamentos: "))
    qtd_faces = int(input("Digite a quantidade de faces do dado: "))

    resultado = lancar_dados(qtd_lancamentos, qtd_faces)

    # Exibe os resultados
    for face, contagem in resultado.items():
        print(f"Face {face}: {contagem} vezes")
```

```python
'''A ACME Inc., uma empresa de 500 funcionários, está tendo problemas de espaço em disco no seu servidor de arquivos. Para tentar resolver este problema, o Administrador de Rede precisa saber qual o espaço ocupado pelos usuários, e identificar os usuários com maior espaço ocupado. Através de um programa, baixado da Internet, ele conseguiu gerar o seguinte arquivo, chamado "usuarios.txt":

<!-- -->

alexandre   456123789
anderson    1245698456
antonio     123456456
carlos      91257581
cesar       987458
rosemary    789456125

Neste arquivo, o nome do usuário possui 15 caracteres. A partir deste arquivo, você deve criar um programa que gere um relatório, chamado "relatório.txt",  no seguinte formato:

ACME Inc.       Uso do espaço em disco pelos usuários
------------------------------------------------------------------------
Nr.  Usuário        Espaço utilizado     % do uso

1    alexandre       434,99 MB             16,85%
2    anderson       1187,99 MB             46,02%
3    antonio         117,73 MB              4,56%
4    carlos       87,03 MB              3,37%
5    cesar         0,94 MB              0,04%
6    rosemary        752,88 MB             29,16%

Espaço total ocupado: 2581,57 MB
Espaço médio ocupado: 430,26 MB

O arquivo de entrada deve ser lido uma única vez, e os dados armazenados em memória, caso sejam necessários, de forma a agilizar a execução do programa. A conversão da espaço ocupado em disco, de bytes para megabytes deverá ser feita através de uma função separada, que será chamada pelo programa principal. O cálculo do percentual de uso também deverá ser feito através de uma função, que será chamada pelo programa principal.'''

from collections import OrderedDict

def bytes_to_megabytes(bytes_value):
    return bytes_value / (1024 * 1024)

def calcular_percentual(uso, total):
    return (uso / total * 100) if total != 0 else 0

def gerar_relatorio(usuarios):
    total_uso = sum(usuarios.values())
    relatorio = []

    relatorio.append("ACME Inc.       Uso do espaço em disco pelos usuários")
    relatorio.append("------------------------------------------------------------------------")
    relatorio.append("Nr.  Usuário        Espaço utilizado     % do uso")

    for i, (usuario, uso) in enumerate(usuarios.items(), start=1):
        uso_mb = bytes_to_megabytes(uso)
        percentual = calcular_percentual(uso, total_uso)
        relatorio.append(f"{i:<4} {usuario:<15} {uso_mb:>12.2f} MB {percentual:>12.2f}%")

    total_mb = bytes_to_megabytes(total_uso)
    media_mb = total_mb / len(usuarios)
    relatorio.append(f"\nEspaço total ocupado: {total_mb:.2f} MB")
    relatorio.append(f"Espaço médio ocupado: {media_mb:.2f} MB")

    return "\n".join(relatorio)

# Leitura do arquivo
usuarios = OrderedDict()
with open("usuarios.txt", "r") as file:
    for line in file:
        parts = line.strip().split()
        if len(parts) == 2:
            usuario = parts[0]
            uso = int(parts[1])
            usuarios[usuario] = uso

# Geração do relatório
relatorio = gerar_relatorio(usuarios)

# Escrevendo em arquivo
with open("relatorio.txt", "w") as file:
    file.write(relatorio)

print("Relatório gerado com sucesso em 'relatorio.txt'.")
```
