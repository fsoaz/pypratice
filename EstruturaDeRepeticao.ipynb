```python
'''Foi feita uma estatística em cinco cidades brasileiras para coletar dados sobre acidentes de trânsito. Foram obtidos os seguintes dados:

    Código da cidade;
    Número de veículos de passeio (em 1999);
    Número de acidentes de trânsito com vítimas (em 1999). Deseja-se saber:
    Qual o maior e menor índice de acidentes de transito e a que cidade pertence;
    Qual a média de veículos nas cinco cidades juntas;
    Qual a média de acidentes de trânsito nas cidades com menos de 2.000 veículos de passeio.'''
codigo_cidade = []
soma_veiculos = []
num_acidentes = []
for i in range(5):
    codigo_cidade.append(int(input("Codigo da cidade: ")))
    veiculos_input = int(input("Veículos: "))
    num_acidentes_input = int(input("Numero de acidentes: "))
    soma_veiculos.append(veiculos_input)
    num_acidentes.append(num_acidentes_input)


#Qual o maior e menor índice de acidentes de transito e a que cidade pertence;
maior_indice = num_acidentes.index(max(num_acidentes))
menor_indice = num_acidentes.index(min(num_acidentes))

print("Cidade com maior indice:", codigo_cidade[maior_indice])
print("Cidade com menor indice:", codigo_cidade[menor_indice])

#Qual a média de veículos nas cinco cidades juntas;
media_veic = sum(soma_veiculos) / len(codigo_cidade)
print("Média de veículos nas cidades:", media_veic)
#Qual a média de acidentes de trânsito nas cidades com menos de 2.000 veículos de passeio.
cidades_com_menos_2000 = [num_acidentes[i] for i in range(len(soma_veiculos)) if soma_veiculos[i] < 2000]
if cidades_com_menos_2000:
    media_acid = sum(cidades_com_menos_2000) / len(cidades_com_menos_2000)
    print("Media de acidente em cidades com menos de 2.000 veículos:", media_acid)
else:
    print("Não há cidades com menos de 2.000 veículos")
```

```python
'''Faça um programa que leia uma quantidade indeterminada de números positivos e conte quantos deles estão nos seguintes intervalos: [0-25], [26-50], [51-75] e [76-100]. A entrada de dados deverá terminar quando for lido um número negativo. '''

while True:
    try:
        numero = float(input("Digite um número:"))
    except ValueError:
        break
    if numero <= 25:
        print("Primeiro intervalo [0-25]")
    elif numero > 25 and numero <= 50:
        print("Segundo intervalo [26-50]")
    elif numero > 50 and numero <= 75:
        print("Terceiro intervalo [51-75]")
    elif numero > 75 and numero > 100:
        print("Quarto intervalo [76-100]")
```

```python
'''O cardápio de uma lanchonete é o seguinte:

    Especificação   Código  Preço
    Cachorro Quente 100     R$ 1,20
    Bauru Simples   101     R$ 1,30
    Bauru com ovo   102     R$ 1,50
    Hambúrguer      103     R$ 1,20
    Cheeseburguer   104     R$ 1,30
    Refrigerante    105     R$ 1,00
    Faça um programa que leia o código dos itens pedidos e as quantidades desejadas. Calcule e mostre o valor a ser pago por item (preço * quantidade) e o total geral do pedido. Considere que o cliente deve informar quando o pedido deve ser encerrado. '''

total_compra = 0
while True:
    precos = {100:1.20, 101:1.30, 102:1.50, 103:1.20, 104:1.30, 105:1.00}
    item = int(input("Digite o código do item: "))
    quantidade = int(input("Insira a quantidade que você deseja: "))

    if item in precos:
        preco = precos[item]
        valor_compra = preco * quantidade
        total_compra += valor_compra
        print("Total a pagar: ", valor_compra)
    else:
        print("Código inválido!")


    continuar = input("Deseja continuar: (s/n)").strip().lower()
    if continuar != 's':
        break

print("Total à pagar: ", total_compra)
```

```python
'''Em uma eleição presidencial existem quatro candidatos. Os votos são informados por meio de código. Os códigos utilizados são:

    1 , 2, 3, 4  - Votos para os respectivos candidatos 
    (você deve montar a tabela ex: 1 - Jose/ 2- João/etc)
    5 - Voto Nulo
    6 - Voto em Branco

    Faça um programa que calcule e mostre:
    O total de votos para cada candidato;
    O total de votos nulos;
    O total de votos em branco;
    A percentagem de votos nulos sobre o total de votos;
    A percentagem de votos em branco sobre o total de votos. Para finalizar o conjunto de votos tem-se o valor zero. '''

jose = 0
joao = 0
maria = 0
lula = 0
nulo = 0
branco = 0
soma_votos = 0
while True:
    try:
        voto = int(input("Digite seu voto (1 - José, 2 - João, 3 - Maria, 4 - Lula, 5 - Nulo, 6 - Branco): "))
        if voto == 1:
            jose += 1
        elif voto == 2:
            joao += 1
        elif voto == 3:
            maria += 1
        elif voto == 4:
            lula += 1
        elif voto == 5:
            nulo += 1
        elif voto == 6:
            branco += 1
        else:
            print("Candidato não localizado")

        soma_votos += 1

    except ValueError:
        print("Entrada inválida, digite novamente")
        continue

    confirmar = input("Finalizar?: (s/n)").strip().lower()

    if confirmar != 'n':
        break


print("Votos Jose: ", jose)
print("Votos João: ", joao)
print("Votos Maria: ", maria)
print("Votos Lula: ", lula)
print("Nulos: ", nulo)
print("Em Branco: ", branco)
print("Total de votos: ", soma_votos)
if soma_votos > 0:
    print(f"Votos nulos (%): { (nulo*100) / soma_votos:.2f}%")
    print(f"Votos em branco (%): { (branco*100) / soma_votos:.2f}%")
else:
    print("Nenhum voto foi registrado")
```

```python
'''Desenvolver um programa para verificar a nota do aluno em uma prova com 10 questões, o programa deve perguntar ao aluno a resposta de cada questão e ao final comparar com o gabarito da prova e assim calcular o total de acertos e a nota (atribuir 1 ponto por resposta certa). Após cada aluno utilizar o sistema deve ser feita uma pergunta se outro aluno vai utilizar o sistema. Após todos os alunos terem respondido informar:

    Maior e Menor Acerto;
    Total de Alunos que utilizaram o sistema;
    A Média das Notas da Turma.

    Gabarito da Prova:

    01 - A
    02 - B
    03 - C
    04 - D
    05 - E
    06 - E
    07 - D
    08 - C
    09 - B
    10 - A

    Após concluir isto você poderia incrementar o programa permitindo que o professor digite o gabarito da prova antes dos alunos usarem o programa. '''

respostas = {1:'A', 2:'B', 3:'C', 4:'D', 5:'E', 6:'E', 7:'D', 8:'C', 9:'B', 10:'A'}
while True:
    nota = 0
    for i in range(1, 11):
        resp = input(f"Resposta {i}: ").upper()
        if resp == respostas[i]:
            nota += 1
        
        
    print("Sua nota: ", nota)

    again = input("Você quer tentar novamente? (s/n): ").lower()
    if again != 's':
        break
```

```python
'''Em uma competição de salto em distância cada atleta tem direito a cinco saltos. No final da série de saltos de cada atleta, o melhor e o pior resultados são eliminados. O seu resultado fica sendo a média dos três valores restantes. Você deve fazer um programa que receba o nome e as cinco distâncias alcançadas pelo atleta em seus saltos e depois informe a média dos saltos conforme a descrição acima informada (retirar o melhor e o pior salto e depois calcular a média). Faça uso de uma lista para armazenar os saltos. Os saltos são informados na ordem da execução, portanto não são ordenados. O programa deve ser encerrado quando não for informado o nome do atleta. A saída do programa deve ser conforme o exemplo abaixo:

Atleta: Rodrigo Curvêllo

Primeiro Salto: 6.5 m
Segundo Salto: 6.1 m
Terceiro Salto: 6.2 m
Quarto Salto: 5.4 m
Quinto Salto: 5.3 m

Melhor salto:  6.5 m
Pior salto: 5.3 m
Média dos demais saltos: 5.9 m

Resultado final:
Rodrigo Curvêllo: 5.9 m'''

salto_ordem = ["Primeiro", "Segundo", "Terceiro", "Quarto", "Quinto"]
salto_list = []
nome_atleta = input("Atleta: ")

for i in range (5):    
    distancia_salto = float(input(f"{salto_ordem[i]} salto: "))
    salto_list.append(distancia_salto)

salto_list.sort()

saltos_restantes = salto_list[1:-1]

media = sum(saltos_restantes) / len(saltos_restantes)

print(f"\nMelhor salto: {max(salto_list)}")
print(f"Pior salto: {min(salto_list)}")
print(f"Media dos demais saltos: {media:.2f}")
print(f"\nResultado final: \n{nome_atleta}: {media:.2f} m")
```

```python
'''Faça um programa que peça um numero inteiro positivo e em seguida mostre este numero invertido.'''

num = int(input("Digite um número: "))
reversed_num = 0

while num != 0:
    digit = num % 10
    reversed_num = reversed_num * 10 + digit
    num //= 10

print("Reversed Number: " + str(reversed_num))
```

```python
'''Faça um programa que mostre os n termos da Série a seguir:

      S = 1/1 + 2/3 + 3/5 + 4/7 + 5/9 + ... + n/m. '''

def serie(term):
    if term <= 0:
        print("Termo deve ser maior que 0")
        return

    a = 1
    b = 1
    soma = 0

    for i in range(1, term+1):
        x = a / b
        soma += x
        print(f"{a} / {b} = {x:.2f}")
        a += 1
        b += 2

    print(f"Soma: {soma:.2f}")

term = int(input("Quantos termos mostrar? "))
serie(term)
```
